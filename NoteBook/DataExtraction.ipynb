```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os

folder = "/content/flipkart_pages"

files = sorted(
    os.listdir(folder),
    key=lambda x: int(x.split("_")[1].split(".")[0])
)

names = []
prices = []
ratings = []
ram = []
display = []
camera = []
battery = []
processor = []

for file in files:

    with open(os.path.join(folder, file), "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    products = soup.find_all("div", {"data-id": True})

    print(file, len(products))

    for i in products:

        try:
            names.append(i.find("div", {"class":"RG5Slk"}).text.strip())
        except:
            names.append(np.nan)

        try:
            prices.append(i.find("div", {"class":"hZ3P6w DeU9vF"}).text.strip())
        except:
            prices.append(np.nan)

        try:
            ratings.append(i.find("span", {"class":"PvbNMB"}).text.strip())
        except:
            ratings.append(np.nan)

        try:
            x = i.find("ul", {"class":"HwRTzP"}).find_all("li")
        except:
            x = []

        try:
            ram.append(x[0].text)
        except:
            ram.append(np.nan)

        try:
            display.append(x[1].text)
        except:
            display.append(np.nan)

        try:
            camera.append(x[2].text)
        except:
            camera.append(np.nan)

        try:
            battery.append(x[3].text)
        except:
            battery.append(np.nan)

        try:
            processor.append(x[4].text)
        except:
            processor.append(np.nan)

df = pd.DataFrame({
    "Name": names,
    "Price": prices,
    "Rating": ratings,
    "RAM": ram,
    "Display": display,
    "Camera": camera,
    "Battery": battery,
    "Processor": processor
})

print(df.shape)
df.head()
```

```python
df.to_csv("flipkart_phones.csv",index=False)
```
