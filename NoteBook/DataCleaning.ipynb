# Dataset Summary

This dataset contains information about 984 smartphones collected from Flipkart, one of India's largest e-commerce platforms. The data was obtained through web scraping using Selenium to access product pages and BeautifulSoup to extract the required information from the HTML.

Each record represents a single smartphone and contains important specifications and details useful for data analysis and machine learning. The dataset includes product names, prices, customer ratings, RAM and storage configurations, display information, camera specifications, battery capacity, and processor details.

The purpose of this dataset is to analyze smartphone features, understand pricing trends, build predictive machine learning models, and develop a recommendation system that suggests suitable smartphones based on user preferences.

#Column Description

**Name**:	Complete name of the smartphone listed on Flipkart.

**Price**: Selling price of the smartphone in Indian Rupees (₹).

**Rating**: Customer rating information available for the product.

**RAM**:RAM and internal storage configuration of the smartphone (e.g., 8 GB RAM | 128 GB Storage).

**Display**:	Display size and screen specifications of the smartphone.

**Camera**:	Rear and front camera specifications of the smartphone.

**Battery**:	Battery capacity of the smartphone measured in mAh.

**Processor**:	Processor or chipset used in the smartphone
