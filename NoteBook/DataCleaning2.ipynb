#Dataset Summary
This dataset contains information about 984 smartphones collected from Flipkart, one of India's largest e-commerce platforms. The data was obtained through web scraping using Selenium to access product pages and BeautifulSoup to extract the required information from the HTML.

Each record represents a single smartphone and contains important specifications and details useful for data analysis and machine learning. The dataset includes product names, prices, customer ratings, RAM and storage configurations, display information, camera specifications, battery capacity, and processor details.

The purpose of this dataset is to analyze smartphone features, understand pricing trends, build predictive machine learning models, and develop a recommendation system that suggests suitable smartphones based on user preferences.

#Column Description
**Name**: Complete name of the smartphone listed on Flipkart.

**Price**: Selling price of the smartphone in Indian Rupees (₹).

**Rating**: Customer rating information available for the product.

**RAM**:RAM and internal storage configuration of the smartphone (e.g., 8 GB RAM | 128 GB Storage).

**Display**: Display size and screen specifications of the smartphone.

**Camera**: Rear and front camera specifications of the smartphone.

**Battery**: Battery capacity of the smartphone measured in mAh.

**Processor**: Processor or chipset used in the smartphone

```python
df["Price"] = (
    df["Price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(int)
)
```

```python
df["Brand"] = df["Name"].str.split().str[0]
```

```python
df["Model"] = df["Name"].str.split().str[1:].str.join(" ")
```

```python
df["Rating"] = df["Rating"].fillna("0 Ratings & 0 Reviews")

df["Rating_Count"] = (
    df["Rating"]
    .str.extract(r'([\d,]+)\s+Ratings')[0]
    .str.replace(",", "", regex=False)
    .fillna("0")
    .astype(int)
)
```

```python
df["RAM_GB"] = (
    df["RAM"]
    .str.extract(r'(\d+)\s*GB\s*RAM')[0]
    .astype(int)
)
```

```python
def extract_rom(text):

    text = str(text)

    if "|" not in text:
        return np.nan

    rom = text.split("|")[1].strip()

    if "TB" in rom:
        return int(rom.split("TB")[0].strip()) * 1024

    elif "GB" in rom:
        return int(rom.split("GB")[0].strip())

    elif "MB" in rom:
        return int(rom.split("MB")[0].strip()) / 1024

    return np.nan

df["ROM_GB"] = df["RAM"].apply(extract_rom).astype(int)
```

```python
df["Color"] = (
    df["Name"]
    .str.extract(r"\(([^,]+),")[0]
)
```

```python
df["Display_cm"] = (
    df["Display"]
    .str.extract(r'([\d.]+)\s*cm')[0]
    .astype(float)
)
```

```python
df["Rear_Camera_MP"] = (
    df["Camera"]
    .str.extract(r'(\d+)MP')[0]
    .astype(float)
)
```

```python
df["Screen_Type"] = "Other"

df.loc[df["Display"].str.contains("AMOLED", case=False, na=False), "Screen_Type"] = "AMOLED"
df.loc[df["Display"].str.contains("OLED", case=False, na=False), "Screen_Type"] = "OLED"
df.loc[df["Display"].str.contains("LCD", case=False, na=False), "Screen_Type"] = "LCD"
```

```python
df["Panel_Type"] = "Other"

df.loc[df["Display"].str.contains("IPS", case=False, na=False), "Panel_Type"] = "IPS"
df.loc[df["Display"].str.contains("LTPO", case=False, na=False), "Panel_Type"] = "LTPO"
```

```python
df["Battery_mAh"] = (
    df["Battery"]
    .str.extract(r'(\d+)\s*mAh')[0]
    .astype(int)
)
```

#Important Note

During data cleaning, it was observed that some product listings contained warranty information instead of processor specifications due to inconsistencies in the scraped data. As a result, the Processor column contained unreliable values for a portion of the dataset. The column was retained in the final dataset for completeness but was excluded from further analysis and model training to maintain data quality.
