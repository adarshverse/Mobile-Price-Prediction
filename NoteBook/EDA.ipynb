```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set_palette("viridis")
```

```python
#Brand Distribution

plt.figure(figsize=(12,6))
df['Brand'].value_counts().plot(kind='bar')
plt.title("Number of Phones by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Price Distribution

plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Number of Phones")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Number of Phones")
plt.show()
```

```python
top=df["Brand"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top.index,y=top.values)
plt.xticks(rotation=45)
plt.title("Top 10 Brands")
plt.show()
```

```python
avg=df.groupby("Brand")["Price"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg.index,y=avg.values)
plt.xticks(rotation=90)
plt.title("Average Price by Brand")
plt.ylabel("Average Price (₹)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="RAM_GB",y="Price")
plt.title("RAM vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="ROM_GB",y="Price")
plt.title("ROM vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="Battery_mAh",y="Price")
plt.title("Battery Capacity vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="Display_cm",y="Price")
plt.title("Display Size vs Price")
plt.show()
```

```python
num=df[
    [
        "Price",
        "Rating_Count",
        "RAM_GB",
        "ROM_GB",
        "Display_cm",
        "Rear_Camera_MP",
        "Battery_mAh"
    ]
]

plt.figure(figsize=(8,6))
sns.heatmap(num.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
